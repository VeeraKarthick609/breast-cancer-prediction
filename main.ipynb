```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
```

```python
df = load_breast_cancer()
features = list(df['feature_names'])
for i in features:
    print(i,end="\n")
```

```python
df['feature_names']
```

```python
dataframe = pd.DataFrame(df['data'])
dataframe.columns = df['feature_names']
dataframe
```

```python
target = df['target']
target_dataframe = pd.DataFrame(target)
target_dataframe.columns = np.array(["Target"])
target_dataframe
```

```python
final_dataframe = pd.concat([dataframe, target_dataframe], axis=1)
final_dataframe
```

```python
#independent dataset
X = dataframe

#dependent datast
Y = target_dataframe

#train_test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=42,train_size=0.8)
```

```python
X_test
```

```python
std_scaler = StandardScaler()

X_train = std_scaler.fit_transform(X_train)
X_test = std_scaler.fit_transform(X_test)
```

```python
model = LinearRegression()

model.fit(X_train, Y_train)
```

```python
reg_predict = model.predict(X_test)
```

```python
##mse and r2 score

mse = cross_val_score(model, X_test, Y_test, scoring='neg_mean_squared_error',cv=10)
mse = np.mean(mse)

r2 = r2_score(reg_predict, Y_test)

print(mse)
print(r2)
```
